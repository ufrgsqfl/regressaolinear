# Uma revisão rápida da regressão linear

```python
import os
import random
import numpy as np
from pylab import mpl, plt
plt.style.use('seaborn')
mpl.rcParams['savefig.dpi'] = 300
mpl.rcParams['font.family'] = 'serif'
os.environ['PYTHONHASHSEED'] = '0'
```

```python
#Gera uma grade uniformemente espaçada de floats para os valores x entre 0 e 10.
x = np.linspace(0, 10)
x
```

```python
#Corrige os valores de semente para todos os geradores de números aleatórios relevantes.
def set_seeds(seed=100):
    random.seed(seed)
    np.random.seed(seed)
set_seeds()
```

```python
#Gera os dados aleatórios para os valores de y
y = x + np.random.standard_normal(len(x))
y
```

```python
#A regressão OLS de grau 1 (ou seja, regressão linear) é criada
reg = np.polyfit(x, y, deg=1)
reg
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'bo', label='data')
plt.plot(x, np.polyval(reg, x), 'r', lw=2.5, label='regressao linear')
plt.legend(loc=0)
```

```python
'''
O intervalo para a variável dependente x é x E[0, 10].
Aumentar o intervalo para, digamos, x E [0, 20] permite “prever”
valores para a variável dependente y além do domínio do conjunto de dados original
por uma extrapolação dados os parâmetros de regressão ótimos
'''
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'bo', label='data')
#Gera um domínio ampliado para os valores x
xn = np.linspace(0, 20)
plt.plot(xn, np.polyval(reg, xn), 'r', lw=2.5, label='regressao linear')
plt.legend(loc=0);
```
